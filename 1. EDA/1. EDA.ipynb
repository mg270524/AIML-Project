# Distribution of Pollutants (Count Plot)

```python
t = np.linspace(-6, 6, 256)
plt.figure(figsize=(15,25))

# Distribution of PM2.5 Avg
plt.subplot(3, 3, 1)
sns.histplot(df['PM2.5 Avg'], kde=True)
plt.title('Distribution of PM2.5 Avg')

# Distribution of PM10_Avg
plt.subplot(3, 3, 2)
sns.histplot(df['PM10_Avg'], kde=True)
plt.title('Distribution of PM10_Avg')

# Distribution of NO2_Avg
plt.subplot(3, 3, 3)
sns.histplot(df['NO2_Avg'], kde=True)
plt.title('Distribution of NO2_Avg')

# Distribution of SO2_Avg
plt.subplot(3, 3, 4)
sns.histplot(df['SO2_Avg'], kde=True)
plt.title('Distribution of SO2_Avg')

# Distribution of CO_Avg
plt.subplot(3, 3, 5)
sns.histplot(df['CO_Avg'], kde=True)
plt.title('Distribution of CO_Avg')

# Distribution of OZONE_Avg
plt.subplot(3, 3, 6)
sns.histplot(df['OZONE_Avg'], kde=True)
plt.title('Distribution of OZONE_Avg')

# Distribution of AQI Value
plt.subplot(3, 3, 7)
sns.histplot(df['AQI Value'], kde=True)
plt.title('Distribution of AQI Value')

# Distribution of Predominant Pollutant
plt.subplot(3, 3, 8)
sns.countplot(data=df, x='Predominant_Parameter', order = df['Predominant_Parameter'].value_counts().index)  #Display in descending order of frequency
plt.title('Distribution of Predominant Pollutant')

plt.show()
```

# Scatterplot (Pollutant vs AQI)

```python
plt.figure(figsize=(15,25))

# PM2.5 Avg vs AQI Value
plt.subplot(3, 3, 1)
sns.scatterplot(data=df, x='PM2.5 Avg', y='AQI Value')
plt.title('PM2.5 Avg vs. AQI Value')

# PM10_Avg vs AQI Value
plt.subplot(3, 3, 2)
sns.scatterplot(data=df, x='PM10_Avg', y='AQI Value')
plt.title('PM10_Avg vs. AQI Value')

# NO2_Avg vs AQI Value
plt.subplot(3, 3, 3)
sns.scatterplot(data=df, x='NO2_Avg', y='AQI Value')
plt.title('NO2_Avg vs. AQI Value')

# SO2_Avg vs AQI Value
plt.subplot(3, 3, 4)
sns.scatterplot(data=df, x='SO2_Avg', y='AQI Value')
plt.title('SO2_Avg vs. AQI Value')

# CO_Avg vs AQI Value
plt.subplot(3, 3, 5)
sns.scatterplot(data=df, x='CO_Avg', y='AQI Value')
plt.title('CO_Avg vs. AQI Value')

# OZONE_Avg vs AQI Value
plt.subplot(3, 3, 6)
sns.scatterplot(data=df, x='OZONE_Avg', y='AQI Value')
plt.title('OZONE_Avg vs. AQI Value')


plt.show()
```

# Boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Predominant_Parameter', y='AQI Value')
plt.title('AQI Value by Predominant Parameter')
plt.show()
```

```python
sns.pairplot(df, vars = ['PM2.5 Avg', 'PM10_Avg', 'NO2_Avg', 'SO2_Avg', 'CO_Avg', 'OZONE_Avg', 'AQI Value'], hue='AQI Value')
```
