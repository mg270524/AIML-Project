```python
aqi = pd.read_csv("aqi3.csv")
```

# Exploratory Data Analysis (EDA) 

```python
sns.countplot(data=aqi, x='Air Quality', order=aqi['Air Quality'].value_counts().index)
plt.title('Distribution of Air Quality Categories')
plt.show()
```

```python
sns.histplot(aqi['Index Value'], kde=True)
plt.title('Distribution of Index Values')
plt.show()
```

```python
plt.figure(figsize=(9, 6))
sns.countplot(data=aqi, x='Prominent Pollutant', order=aqi['Prominent Pollutant'].value_counts().index)
plt.title('Distribution of Prominent Pollutants')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data=aqi, x='Air Quality', y='Index Value')
plt.title('Air Quality vs. Index Value')
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.boxplot(data=aqi, x='Prominent Pollutant', y='Index Value')
plt.title('Prominent Pollutant vs. Index Value')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(25, 200))
sns.countplot(data=aqi, y='City', hue='Air Quality', order=aqi['City'].value_counts().index)
plt.title('City-wise Air Quality Distribution')
plt.show()
```

```python
sns.pairplot(aqi)
plt.show()
```
